# Classification

### Determining customer churn

One of the most common problems in business. Determining the mood of the customer.

If a business owner can predict in advance that a customer will not return, will not use a product or service again, he can take steps to retain the customer, to change his mind.

In this practice, we predict whether a customer will stay or not by analyzing the data of online store customers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, RocCurveDisplay, auc
```

# Preparing Data for ML

```python
print(f"Number of non-existent rows: {df.isnull().sum().sum()}")
missing_rows = df[df.isnull().any(axis=1)].shape[0]
print(f"Number of non-existent columns: {missing_rows}")
print(f"Number of non-existent columns: % {np.round(missing_rows/len(df)*100)}")
```

33% is a lot. But do we use all the columns?

Let's select the ones with high correlation and some categorical columns

```python
data = df[['Churn','Tenure','Complain','DaySinceLastOrder','CashbackAmount','MaritalStatus','Gender']]
```

```python
print(f"Number of non-existent rows: {data.isnull().sum().sum()}")
missing_rows = data[data.isnull().any(axis=1)].shape[0]
print(f"Number of non-existent columns: {missing_rows}")
print(f"Number of non-existent columns: % {np.round(missing_rows/len(data)*100)}")
```

10% of rows value is not available.
We remove null values from table.

```python
data = data.dropna()
data.shape
```

```python
data['Churn'].value_counts()/len(data)*100
```

# ML

```python
# Converting text columns to numbers
encoded = pd.get_dummies(data)
encoded.head()
```

```python
X = encoded.drop('Churn', axis=1)
y = encoded['Churn']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

# Logistic Regression

```python
# Training Model
LR_model = LogisticRegression()
LR_model.fit(X_train, y_train)

# Evaluating Model
y_pred = LR_model.predict(X_test)
print(classification_report(y_test, y_pred))
print(f"Model accuracy: {accuracy_score(y_test, y_pred)}")

# Confusion Matrix
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True, fmt="g")
plt.show()

## ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

# Support Vectore Machines

```python

# Training Model
svm_model = SVC()
svm_model.fit(X_train, y_train)

# Evaluating Model
y_pred = svm_model.predict(X_test)
print(classification_report(y_test, y_pred))
print("Model accuracy:", accuracy_score(y_test,y_pred))

## Confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()

## ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

# Decision Tree

```python
# Training Model
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)

# Evaluating Model
y_pred = tree_model.predict(X_test)
print(classification_report(y_test, y_pred))
print("Model accuracy:", accuracy_score(y_test,y_pred))

## Confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()

## ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

# Random Forest

```python
# Training Model
RF_model = RandomForestClassifier(n_estimators=9)
RF_model.fit(X_train, y_train)

# Evaluating Model
y_pred = RF_model.predict(X_test)
print(classification_report(y_test, y_pred))
print("Model accuracy:", accuracy_score(y_test,y_pred))

# Confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()

# ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```
